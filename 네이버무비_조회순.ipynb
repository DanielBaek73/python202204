```python
import requests                       #웹페이지 연결
from bs4 import BeautifulSoup as bs   #가져온 페이지를 분석하기 용이하게 정제
import pandas as pd   
html = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cnt&date=20220507')
soup=bs(html.text)
soup

```

```python
len(soup.select('tr'))  #태그를 선택할때는 soup.select()을 사용
# len(soup.select('tbody > tr'))  
```

```python
# len(soup.select('tr'))  #태그를 선택할때는 soup.select()을 사용
len(soup.select('tbody > tr'))  
```

```python
len(soup.select('tbody > tr > td.title'))  
```

```python
movies = soup.select('tbody > tr > td.title')
movie = movies[0]
```

```python
movies[0].select('a')[0].text
```

```python
movie_data = []
rank = 1

movies = soup.select('tbody > tr > td.title')

for movie in movies:
  title = movie.select('a')[0].text


  movie_data.append(['조회순',rank,title])
  rank += 1

movie_data
```

```python
df = pd.DataFrame(movie_data,columns = ['영화랭킹','순위','타이틀'])
df
```

```python
df.to_excel('네이버_무비_조회순.xlsx',index=False)
```
